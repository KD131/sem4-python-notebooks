```python
%matplotlib inline
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
student_attendance = {'day1':33, 'day2':34,'day3':29,'day4':31,'day5':28,'day6':26,'day7':30}
days = list(student_attendance.keys())
nums = list(student_attendance.values())
plt.figure()
plt.title("Student attendance", fontsize=18)
plt.xlabel("Days", fontsize=16)
plt.ylabel("Attendance", fontsize=16)
plt.axis(ymin=0, ymax=max(nums) + 5)
plt.tick_params(axis="both", labelsize=14)
plt.plot(days, nums)
```

```python
data = {'apple': 10,'banana':2,'citrus':32,'dragon fruit':8}
fig1, ax1 = plt.subplots()
ax1.pie(data.values(), labels=data.keys(), autopct=lambda p:"{:.2f}% ({:.2f})".format(p,(p/100)*sum(data.values())), shadow=True, startangle=90)
ax1.set_aspect("equal")
ax1.legend(data.keys(), loc="right")
```

```python
# 1.10
percentage_range = range(0, 101, 10)
count_per_percentage = []
percentages = [min(s.get_total_progress()//10*10, 100) for s in students] # rounds down to nearest 10, max 100
for p in percentage_range:
    count_per_percentage.append(percentages.count(p))
    
plt.figure()
plt.bar(percentage_range, count_per_percentage, width=5)
plt.title("Percentage way through school", fontsize=18)
plt.xlabel("%", fontsize=16)
plt.ylabel("No. of students", fontsize=16)
plt.tick_params(axis="both", labelsize=12)
```

```python
# 3.1

def percentages_pie(students):
    #copy-paste from above since it's the same data.
    percentage_range = range(0, 101, 10)
    # rounds down to nearest 10, max 100
    percentages = [min(s.get_total_progress()//10*10, 100) for s in students]
    # ignores category with 0 count
    count_per_percentage = {p:percentages.count(p) for p in percentage_range if percentages.count(p)}
        
    fig1, ax1 = plt.subplots()
    ax1.pie(count_per_percentage.values(), labels=count_per_percentage.keys(), counterclock=False, autopct=lambda p:"{:.2f}%\n({:.0f})".format(p,(p/100)*sum(count_per_percentage.values())), shadow=True, startangle=90)
    ax1.set_aspect("equal")
    ax1.legend(count_per_percentage.keys(), loc="best")
    
percentages_pie(students)
```

```python
# 3.2

def students_per_course_bar(students):
    # this can only plot courses that have students in them, as I get the list of courses from the students.
    data = {}
    for s in students:
        for c in s.data_sheet:
            data.setdefault(c.name, 0)
            data[c.name] += 1
    
    
    plt.figure()
    plt.bar(data.keys(), data.values(), width=0.5)
    plt.title("Students per course", fontsize=18)
    plt.xlabel("Courses", fontsize=16)
    plt.ylabel("No. of students", fontsize=16)
    plt.axis(ymin=0, ymax=len(students)) # shows out of total no. of students. optional.
    plt.tick_params(axis="both", labelsize=12)
    plt.xticks(rotation=45, horizontalalignment="right")
    
students_per_course_bar(students)
```
