```python
%matplotlib inline
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
student_attendance = {'day1':33, 'day2':34,'day3':29,'day4':31,'day5':28,'day6':26,'day7':30}
days = list(student_attendance.keys())
nums = list(student_attendance.values())
plt.figure()
plt.title("Student attendance", fontsize=18)
plt.xlabel("Days", fontsize=16)
plt.ylabel("Attendance", fontsize=16)
plt.axis(ymin=0, ymax=max(nums) + 5)
plt.tick_params(axis="both", labelsize=14)
plt.plot(days, nums)
```

```python
data = {'apple': 10,'banana':2,'citrus':32,'dragon fruit':8}
fig1, ax1 = plt.subplots()
ax1.pie(data.values(), labels=data.keys(), autopct=lambda p:"{:.2f}% ({:.2f})".format(p,(p/100)*sum(data.values())), shadow=True, startangle=90)
ax1.set_aspect("equal")
ax1.legend(data.keys(), loc="right")
```
