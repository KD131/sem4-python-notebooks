```python
%matplotlib inline
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
student_attendance = {'day1':33, 'day2':34,'day3':29,'day4':31,'day5':28,'day6':26,'day7':30}
days = list(student_attendance.keys())
nums = list(student_attendance.values())
plt.figure()
plt.title("Student attendance", fontsize=14)
plt.xlabel("Days", fontsize=10)
plt.ylabel("Attendance", fontsize=10)
plt.axis(ymin=0, ymax=max(nums) + 5)
plt.tick_params("both", )
plt.plot(days, nums)
```
